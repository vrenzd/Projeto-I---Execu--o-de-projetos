- desenvolvimento da árvore de decisão para a base de conhecimento "aceitar o emprego", conforme exemplo em aula.

- Importanto as bibliotecas

```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
```

- Definição da base de conhecimento em pandas

```python
df_treino=pd.DataFrame({
    'salario':['alto','baixo','baixo','alto','alto','baixo'],
    'localizacao':['longe','perto','longe','longe','perto','longe'],
    'funcao':['interessante','desinteressante',
              'interessante','desinteressante',
              'interessante','desinteressante'],
    'decisao':['sim','nao',
               'sim','nao',
               'sim','nao']
    })

print(df_treino)
```

- Divisão da base de conhecimento em dados: 
- **X** (colunas de variáveis)  
- **Y** (coluna decisória)

```python
x_treino=df_treino.iloc[:,:-1]
y_treino=df_treino.iloc[:, -1]

print(y_treino)
```

- Variáveis numéricas para substituição na base

```python
alto=1
baixo=0
perto=1
longe=0
interessante=1
desinteressante=0
```

- Preparando a base de conhecimento conforme os parâmetros 
da biblioteca sklearn (a biblioteca não aceita variáveis linguísticas)

```python
x_treino=x_treino.replace({
    'alto':alto, 'baixo':baixo,
    'perto':perto, 'longe':longe,
    'interessante':interessante, 
    'desinteressante':desinteressante
})
```

- Montando a árvore

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
model.fit(x_treino, y_treino)
```

- comando para ver como a árvore foi montada pelo sistema

```python
plt.show(tree.plot_tree(model))
```
