desenvolvimento da árvore de decisão para a base de conhecimento "aceitar o emprego", conforme exemplo em aula.

```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt

df_treino=pd.DataFrame({
    "dor no peito": ['N', 'N', 'S','S', 'S', 'N'],
    'Peso':[90, 70, 80,70, 90, 80],
    'atack cardigan': ['S', 'N', 'S','S', 'S', 'N']
})
print(df_treino)
```

```python
x_treino=df_treino.iloc[:,:-1]
y_treino=df_treino.iloc[:,-1]
```

```python
x_treino=x_treino.replace({'N':0,'S':1})
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
model.fit(x_treino,y_treino)
```

```python
#comando para ver como a árvore foi montada pelo sistema
plt.show(tree.plot_tree(model))
```

```python
df_test=pd.DataFrame({
    "dor no peito": ['N', 'N', 'S','S'],
    'Peso':[65,85,75,50],
    'atack cardigan': ['S', 'N', 'S','S']
})
print(df_treino)
```

```python
x_test=df_test.iloc[:,:-1]
y_test=df_test.iloc[:,-1]
```

```python
x_test=x_test.replace({'N':0,'S':1})
```

```python
y_pred=model.predict(x_test)
```

```python
print(f'avaliação de arvore',y_pred)
```

```python
print(metrics.confusion_matrix(y_test,y_pred))
```
